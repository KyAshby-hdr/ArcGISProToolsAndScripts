```python
import requests
import pandas as pd
import requests
import ipywidgets as widgets
from ipywidgets import Box
states = ["AL", "AK", "AZ", "AR", "CA", "CO", "CT", "DC", "DE", "FL", "GA", "HI", "ID", "IL", "IN", "IA", "KS", "KY", "LA", "ME", "MD", "MA", "MI", "MN", "MS", "MO", "MT", "NE", "NV", "NH", "NJ", "NM", "NY", "NC", "ND", "OH", "OK", "OR", "PA", "RI", "SC", "SD", "TN", "TX", "UT", "VT", "VA", "WA", "WV", "WI", "WY"]
state_dropdown = widgets.Dropdown(
    options=states,
)
start_date_widget = widgets.DatePicker()
end_date_widget = widgets.DatePicker()
widget_list = [
    state_dropdown,
    start_date_widget,
    end_date_widget
]
widget_box = Box(
    children = widget_list
)
widget_box
```

```python
state_cd = state_dropdown.value
start_date = start_date_widget.value
end_date = end_date_widget.value
url = f"https://waterservices.usgs.gov/nwis/dv/?format=json&stateCd={state_cd}&siteStatus=active&siteType=ST"
response = requests.get(url)
gage_list = []
if response.status_code == 200:
    data = response.json()
    for series in data["value"]["timeSeries"]:
        gage_name = series["sourceInfo"]["siteName"]
        gage_id = series["sourceInfo"]["siteCode"][0]["value"]
        gage_list.append((f"{gage_name}: {gage_id}",gage_id))
usgs_gage_dropdown = widgets.Dropdown(
    options = gage_list
)
usgs_gage_dropdown
```

```python

usgs_gage_id = usgs_gage_dropdown.value
usgs_gage_file_url = f"https://waterservices.usgs.gov/nwis/dv?format=json&siteStatus=all&site={usgs_gage_id}&agencyCd=USGS&statCd=00003&parameterCd=00060&startDT={start_date}&endDT={end_date}"
response = requests.get(usgs_gage_file_url)
if response.status_code == 200:
    data = response.json()
```

```python
df = pd.DataFrame(columns=["DateTime", "DailyAverage", "Qualifier"])
for value in data["value"]["timeSeries"][0]["values"][0]["value"]:
    date_time = value["dateTime"]
    daily_average = float(value["value"])
    qualifier = value["qualifiers"]
    new_df_row = pd.DataFrame({
        "DateTime": [date_time],
        "DailyAverage": [daily_average],
        "Qualifier": [qualifier]
    })
    df = pd.concat([df, new_df_row], ignore_index=True)
df
```

```python
from arcpy import mp, sa
import time
aprx = mp.ArcGISProject("CURRENT")

#! MAPS
extents_map = aprx.listMaps("Extents")[0]
zoomed_map = aprx.listMaps("Zoomed")[0]

#! LAYOUTS
extents_layout = aprx.listLayouts("HDR_Modern*Extents")[0]
zoomed_layout = aprx.listLayouts("HDR_Modern*Zoomed")[0]

```

```python
#! EXTENTS MAP

#! TEXT ELEMENTS
project_name_text = extents_layout.listElements("TEXT_ELEMENT", "Title 1")[0]
scenario_and_flow_text = extents_layout.listElements("TEXT_ELEMENT", "Title 2")[0]
variable_text = extents_layout.listElements("TEXT_ELEMENT", "Figure Number")[0]
legend = extents_layout.listElements("LEGEND_ELEMENT")[0]

project_name_text.text = "Eckert Road Bridge"
scenarios_list = ["Existing","Proposed"]
variables_list = ["Depth", "WSE", "Velocity"]
flow_list = ["2year","10year","50year","100year","500year"]
scenario_len = len(scenarios_list)
variables_len = len(variables_list)
flow_len = len(flow_list)
total_len = scenario_len * variables_len * flow_len
for lyr in extents_map.listLayers():
    if (lyr.name != "World Imagery") and (lyr.name != "Hybrid Reference Layer"):
        lyr.visible = False
time.sleep(2)
lyr_list = extents_map.listLayers()
pdf_count = 1
for scenario in scenarios_list:
    for variable in variables_list:
        for flow in flow_list:
            for lyr in lyr_list:
                if lyr.longName == scenario:
                    lyr.visible = True
                if lyr.longName == f"{scenario}\\{variable}":
                    lyr.visible = True
                if lyr.longName == f"{scenario}\\{variable}":
                    lyr.visible = True
                if lyr.longName == f"{scenario}\\{variable}\\{flow}":
                    lyr.visible = True
                    if variable == "Depth":
                        variable_text.text = "Water Depth"
                        legend.title = "DEPTH (FT)"
                    elif variable == "WSE":
                        for lyr in lyr_list:
                            if lyr.longName == f"{scenario}\\{variable}\\{flow}_Contour":
                                lyr.visible = True
                        variable_text.text = "Water Surface Elevation"
                        legend.title = "WATER SURFACE ELEVATION (FT)"
                    elif variable == "Velocity":
                        for lyr in lyr_list:
                            if lyr.longName == f"{scenario}\\{variable}\\{flow}_Vector":
                                lyr.visible = True
                        variable_text.text = "Water Velocity and Direction"
                        legend.title = "VELOCITY (FT/S)"
                    if flow == "2year":
                        flow_text_insert = "2-year"
                    elif flow == "10year":
                        flow_text_insert = "10-year"
                    elif flow == "50year":
                        flow_text_insert = "50-year"
                    elif flow == "100year":
                        flow_text_insert = "100-year"
                    elif flow == "500year":
                        flow_text_insert = "500-year"
                    if scenario == "Existing":
                        scenario_and_flow_text.text = f"Pre-Project Conditions, {flow_text_insert} flow"
                    else:
                        scenario_and_flow_text.text = f"{scenario} Conditions, {flow_text_insert} flow"
                    extents_layout.exportToPDF(f"C:\\Users\\kyashby\\Documents\\PROJECTS\\GIS_REQUESTS\\Eckert_FigureHelp_JuliaLindgren\\7.2_WIP\\exports\\Extents\\{scenario}_{variable}_{flow}_Extents.pdf")
                    print(f"Successfully exported {scenario}_{variable}_{flow}_Extents.pdf, {pdf_count} of {total_len}")
                    pdf_count += 1
            for lyr in extents_map.listLayers():
                if (lyr.name != "World Imagery") and (lyr.name != "Hybrid Reference Layer"):
                    lyr.visible = False
```
